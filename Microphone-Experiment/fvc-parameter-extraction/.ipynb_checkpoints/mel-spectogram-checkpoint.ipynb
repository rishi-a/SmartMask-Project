Mel Spectogram For Audio Files

```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
```

```python
import os
mypath="clothdata/"
(_, _, filenames) = next(os.walk(mypath))
```

```python
from thinkdsp import read_json
wave = read_json(filepath=mypath+filenames[0])
```

### Mel Filter Banks


```python
filter_banks = librosa.filters.mel(n_fft=2048, sr=22050, n_mels=10)
```

```python
filter_banks.shape
#(number of mel Bands X n_fft/2+1)
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(filter_banks, sr=sr, x_axis="linear")
plt.colorbar(format="%+2.f")
plt.show()
```

### Extracting Mel Spectogram

```python
sr = 16000
mel_spectrogram = librosa.feature.melspectrogram(wave.ys, sr=sr, n_fft=2048, hop_length=512, n_mels=90)
```

```python
mel_spectrogram.shape
#(Number of mel-bands X Number of temporal bins extracted from the signal)
```

```python
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(log_mel_spectrogram, 
                         x_axis="time",
                         y_axis="mel", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```
