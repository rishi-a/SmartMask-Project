```python
import pandas as pd
import numpy as np
import cbor2
```

```python
def cbor2csv(filename):
    '''
    Function to convert cbro files to csv.
    Depends on: Numpy, Pandas and Cbor2 library
    Input: Filename with cbor
    Output: Stores the same name csv file
    '''
    with open(filename, 'rb') as fp:
        obj = cbor2.load(fp)  
        pd.DataFrame(data=np.array(obj['payload']['values']),columns=['X','Y','Z']).to_csv(filename.split('.')[0]+".csv", index=False)

```

```python
df = pd.read_csv('temperature.csv')
df.index = pd.to_datetime(df['timestamp'],unit='ms')
df = df.drop(['timestamp'], axis=1)
df = df.rename(columns={'bluetooth_le-CE:C9:CF:DE:90:5A-Temperature-1':'Temperature'})
ax = df.plot(figsize=(10,3))
ax.set_xlabel('Time') 
ax.set_ylabel('Temperature in C');
print("Median Time Difference Between Samples = {} seconds".format(df.index.to_series().diff().median().microseconds*10e-6))
```
